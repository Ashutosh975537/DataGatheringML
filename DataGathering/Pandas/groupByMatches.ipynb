# The problem statement is lets create a funtion that will give the maximum runs scored by a batmans against anyteam

```python
def getMaximumRunsScoredAgainst(player):
    playerMask = delivery['batsman'] == player 
    playerDF = delivery[playerMask] 
    return playerDF.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending = False).index[0]
```

```python
 # isin
# problem statement is who is the most destructive bastman in the History of IPL in the death
# overs

delivery.head()

def getDestructiveInCertainOver():
    over16Mask = delivery['over'] == 16 
    over17Mask = delivery['over'] == 17
    over18Mask = delivery['over'] == 18 
    over19Mask = delivery['over'] == 19 
    over20Mask = delivery['over'] == 20 

    combineDeathMask = over16Mask | over17Mask | over18Mask | over19Mask | over20Mask


    deathOverDf = delivery[combineDeathMask]
    runsScored = deathOverDf.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False)
    ballsPlayed = deathOverDf.groupby('batsman').size().sort_values(ascending = False) 

    strikeRate = []
    for index in runsScored.index :
        if runsScored[index] < 500 :
            break 
        strikeRate.append([index,runsScored[index]/ballPlayed[index]*100]);

    strikeRate.sort()
    print(strikeRate)
```

```python
getDestructiveInCertainOver()
```

# the corr function or simply corelations it is highly used in the machine learnings

```python
# it tells us about how two things are varying with respect to each other
# for understanding corr we must have the knowledge of the varience and standard deviiation
# the corr function does calculations only on the numerical data it will give us pivot table
numericalData = matches[['id','season','dl_applied','win_by_runs','win_by_wickets']]
pt = numericalData.corr()
```

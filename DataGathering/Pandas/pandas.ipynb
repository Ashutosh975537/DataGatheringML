```python
# creating a function that will give the city count matches
def get_city_matches(city):
    mask = df['city'] == city;
    return df[mask].shape[0];
```

```python
matchesCount = get_city_matches('Chennai') ;
print(matchesCount);
```

```python
# it might be possible that there are for that we need to create multiple mask
# note that whenever we want to do and of the two boolean values we use & and not and 
# for every particular values we will be creating a mask and we will bw asking ourselves 
# if it is and condition or it is or consition
cityMask = df['city'] == 'Hyderabad' ;
dataMask = df['date'] > '2010-01-01' ;
print(df[cityMask & dataMask].shape);

def get_city_date(city , date):
    cityMask = df['city'] == city ;
    dataMask = df['date'] > date ;
    print(df[cityMask & dataMask]);
    
```

```python
get_city_date('Chennai','2011-01-01');
```

```python
# problem statement we need find how many team have played how many matches 
def getTotalMatchesPlayed(team):
    teamOneMask = df['team1'] == team ;
    teamTwoMask = df['team2'] == team ;
    print(df[teamOneMask | teamTwoMask].shape[0]) ;
    
```

```python
getTotalMatchesPlayed('Royal Challengers Bangalore');
getTotalMatchesPlayed('Kochi Tuskers Kerala')
```

# sort values method Here we will apply the sort values method

```python
# it is applicable both on the series and  the dataframes and 
# it is highly usefull
totalMatchesPlayed = teamOneMatchesCount + teamTwoMatchesCount ;
print(totalMatchesPlayed);
# If we to sort this thing on the basis of the values we can simply do
```

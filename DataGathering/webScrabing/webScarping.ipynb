```python
# It will create the formatting like the HTML
soup = BeautifulSoup(webPage,'lxml')
print(soup)
```

```python
allRepositeries = soup.find_all('div',class_ = 'Box')
allRepositeries
```

```python
repoTitle = []
allRepoTitle = soup.find_all('span',class_ = 'repo')
for repos in allRepoTitle :
    repoTitle.append(repos.text.strip())
repoTitle
```

```python
repoTypes = []
allRepoType = soup.find_all('span',class_ = 'Label')
allRepoType
for repoType in allRepoType :
    repoTypes.append(repoType.text.strip()[:6])
repoTypes
```

```python
allDescription = soup.find_all('p',class_ = 'pinned-item-desc color-fg-muted text-small d-block mt-2 mb-3')
description = []

for desc in allDescription :
    description.append(desc.text.strip())
description
```

```python
allTech = soup.find_all('span',itemprop = 'programmingLanguage')
techUsed = []
for tech in allTech:
    techUsed.append(tech.text.strip())
techUsed
```

```python
# creating df from list we can pass a dictionary of the key value pairs such as

df = pd.DataFrame()
df
```

```python

def getUserGitHubData(userName):
    global df
    owner = []
    title = []
    repoType = []
    tech = []
    description = []
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
    url = 'https://github.com/{}'.format(userName)
    webPage = req.get(url,headers = headers).text
    if(webPage == 'Not Found'):
        return 
    soup = BeautifulSoup(webPage,'lxml')
    container = soup.find_all('div',class_ = 'Box')
    for data in container :
        owner.append(userName)
        repoType.append(data.find('span',class_ = 'Label').text.strip() if data.find('span', class_='Label') else '-')
        title.append(data.find('span',class_ = 'repo').text.strip() if data.find('span', class_='repo') else '-')
        description.append(data.find('p',class_ = 'pinned-item-desc color-fg-muted text-small d-block mt-2 mb-3').text.strip() if data.find('span', class_='pinned-item-desc color-fg-muted text-small d-block mt-2 mb-3') else '-')
        tech.append(data.find('span', itemprop='programmingLanguage').text.strip() if data.find('span', itemprop='programmingLanguage') else '-')
    print(owner , title , repoType , tech ,description)
    d = {'owner' : owner , 'title' : title , 'type' : repoType,'tech' : tech , 'description' : description}
    temp_df = pd.DataFrame(d)
    print(temp_df)
    df = pd.concat([df, temp_df], ignore_index=True)
        
        
    
```

```python
getUserGitHubData('akshatsoni1')
df
```

```python
df.to_csv('gitHub.csv')
```
